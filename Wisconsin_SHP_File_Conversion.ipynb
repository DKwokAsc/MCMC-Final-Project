#**SHP Conversion File Part**

Assumptions:
- wi_2024_gen_prec.zip is inputted in the main files folder (not in sample_data)
- ZIP file is extracted with all necessary files (dbf, shp and shx at the minimum)

Notes:
- It is completely normal to see that there are several warnings as some columns are filled with NA values. (To decide whether to keep them or exclude them).

- Skip this part if wi_2024_gen_prec.json is already created in the file directory

- File logic will be used to save a copy of the graph later on

- Download time should take about a minute

```python
# Find and locate the shp file in the data located in the main directory in files
shp_path = "./wi_2024_gen_prec.shp"

# Create a graph and nodes for JSON using the shp file
graph = Graph.from_file(shp_path)

# Safety net to check file has no problems
print(graph.nodes[0])
print(len(graph.nodes))

# Export selected file to download_path directory
download_path = "./wi_2024_gen_prec_graph.json"

# Convert file from shp to JSON
graph.to_json(download_path)

# Double check file has been saved
print(f"Saved to: {download_path}")
```
